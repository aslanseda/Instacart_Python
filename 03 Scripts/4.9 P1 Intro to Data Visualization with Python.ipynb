# 3. Cleaning Customer dataset

```python
# Step 4 - Wrangle the data

df_customers = df_customers.drop(columns = {'First Name'}, axis=1)
df_customers = df_customers.drop(columns = {'Surnam'}, axis=1)

df_customers.rename(columns = {'Gender': 'gender'}, inplace = True)
df_customers.rename(columns = {'STATE': 'state'}, inplace = True)
df_customers.rename(columns = {'Age': 'age'}, inplace = True)
df_customers.rename(columns = {'n_dependants': 'numbers_of_dependants'}, inplace = True)
df_customers.rename(columns = {'fam_status': 'family_status'}, inplace = True)
```

I drop First name and Surname columns as they are useless for my analysis. Then I rename columns to have a stantard naming and have better understanding like n_dependants to numbers_of_dependants.

# 4. Checking for duplicates, nulls and mixed-types in Customer dataset

```python
# Step 5 - data quality and consistency checks
df_customers.dtypes
```

```python
# I change the data type of user_id to str as it is not a quantitative
df_customers['user_id'] = df_customers['user_id'].astype('str')
```

# 5. Checking data types

```python
#Checking for mixed types in columns
for col in df_customers.columns.tolist():
    weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_customers[weird]) > 0:
        print (col + ' / mixed')
    else:
        print (col + ' / not-mixed')
```

```python
# checking the duplicates
df_dups = df_customers[df_customers.duplicated()]
```
