# 1. Importing Libraries

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# 2. Importing Data Frame

```python
# Creating the path for the data
path = r'C:\Users\sdaon\Documents\Instacart Basket Analysis'
```

```python
# Importing orders_products_merged
df = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_customers_combined.pkl'))
```

```python
df.head()
```

# 3. Create bar chart

```python
# create a bar chart using orders day of week
bar = df['orders_day_of_week'].value_counts().plot.bar()
plt.title('Frequency of Orders per Day')
plt.xlabel('Day')
plt.ylabel('Frequency (in millions)')
```

```python
# export the bar chart
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_ords_dow.png'))
```

# 8. Histogram about Orders by hour of the day

```python
hist_hour = df['order_hour_of_day'].plot.hist(bins = 24)
```

# 9. Bar chart about Loyalty flag

```python
# Create bar chart for customer loyalty
cust_bar = df['loyalty_flag'].value_counts().plot.bar()
plt.xlabel('Distribution of Products Ordered',fontsize=12, fontweight = 'bold')
plt.ylabel('Type of Customer',fontsize=12, fontweight = 'bold')
```

## This indicates that regular consumers place the most orders, followed by loyal customers.

```python
# Export a bar chart
cust_bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_loyalty.png'))
```

# 10. Line chart about prices and Orders by hour of the day

```python
# I crate a small data frame for price and order hour of day
df_3 = small[['prices', 'order_hour_of_day'] ]
```

```python
# Crate a line chart
line_price = sns.lineplot(data = df_3, x = 'order_hour_of_day' , y = 'prices')
plt.xlabel('Hour of the Day')
plt.ylabel('Average of the Prices of items')
plt.title('Average Prices of Items, by the Hour of the Day')
```

```python
# Export a line chart
line_price.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_price_ords_hour.png'))
```

# 11. Line chart about age and number of dep

```python
# I crate a small data frame about age and number of depants
df_4 = small[['age', 'numbers_of_dependants']]
```

```python
# Crate a line chart
line_age = sns.lineplot (data = df_4, x = 'age', y = 'numbers_of_dependants')
plt.xlabel('Age')
plt.ylabel('Children')
plt.title('Age and Number of Children')
```
